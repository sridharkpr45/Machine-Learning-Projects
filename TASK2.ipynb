DATA CLEANING

```python
# Columns Description
def dataDescription():
    list_item = []
    for col in df_display.columns :
        list_item.append([col,df_display[col].dtype, df_display[col].isnull().sum(),round(df_display[col].isnull().sum()/len(df_display)*100,2)
                          ,df_display[col].nunique(),list(df_display[col].drop_duplicates().sample(4).values)])
    Descdata = pd.DataFrame(data = list_item, columns = ['Column','Data Type', 'Missing Value','Pct Missing Value', 'Num Unique', 'Unique Sample'])
    return Descdata
dataDescription()
```

```python
#remove missing values from the dataframe df_display.
df_display = df_display.dropna()
df_display
```

```python
df_display.shape
```

```python
# Renaming the Columns
df_display = df_display.rename(columns={'Restaurant ID': 'restaurant_id'})
df_display = df_display.rename(columns={'Restaurant Name': 'restaurant_name'})
df_display = df_display.rename(columns={'Cuisines': 'cuisines'})
df_display = df_display.rename(columns={'Aggregate rating': 'aggregate_rating'})
df_display = df_display.rename(columns={'Votes': 'votes'})
df_display
```

```python
# Cross Tabulate Restaurant Name and Cuisines
TabRestaurant_nameCuisines = pd.crosstab(df_display['restaurant_name'],df_display['cuisines'])
TabRestaurant_nameCuisines
```

```python
# Checking on restaurant name value
TabRestaurant_nameCuisines.loc['feel ALIVE'].values
```

```python
# Restaurant Names Sample
df_display['restaurant_name'].sample(25, random_state=101)
```

```python
# Measure Similarity
# The jaccard_score is a metric used to measure the similarity between two sets.It's coefficient ranges from 0 to 1. 
print(jaccard_score(TabRestaurant_nameCuisines.loc["Olive Bistro"].values, TabRestaurant_nameCuisines.loc["Rose Cafe"].values))
```

```python
# Create Similarity Value DF
jaccardDist = pdist(TabRestaurant_nameCuisines.values, metric='jaccard')
jaccardMatrix = squareform(jaccardDist)
jaccardSim = 1 - jaccardMatrix
dfJaccard = pd.DataFrame(jaccardSim,index=TabRestaurant_nameCuisines.index,columns=TabRestaurant_nameCuisines.index)
dfJaccard
```

```python
df_display['restaurant_name'].sample(25)
```

 Final Recommendation System

```python
# Input Initial Restaurant Name
resto = 'The Crab Shack'
sim = dfJaccard.loc[resto].sort_values(ascending=False)
sim = pd.DataFrame({'restaurant_name': sim.index, 'simScore': sim.values})
sim = sim[(sim['restaurant_name']!= resto) & (sim['simScore']>=0.5)].head(5)
# Merge The Rating
RestoRec = pd.merge(sim,df_display[['restaurant_name','aggregate_rating']],how='inner',on='restaurant_name')
FinalRestoRec = RestoRec.sort_values('aggregate_rating',ascending=False).drop_duplicates('restaurant_name',keep='first')
FinalRestoRec
```
